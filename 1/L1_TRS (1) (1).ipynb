```python
#1-3 scipy
import numpy as np
from scipy.integrate import odeint

def system(y, t):
    x, v = y
    dxdt = v
    dvdt = -4/9.1 * (2*x + 2*x*np.tan(x**2)**2)
    return [dxdt, dvdt]

# Начальные значения
y0 = [0, 0.09]

# Временные точки, для которых нужно решить систему
t = np.linspace(0, 10, 10000)

# Решение системы
sol = odeint(system, y0, t)

# Получение результатов
x = sol[:, 0]
v = sol[:, 1]

plt.plot(t,x)
plt.plot(t,v)
```

```python
#4 задание scipy
import numpy as np
from scipy.integrate import solve_bvp
import matplotlib.pyplot as plt

# Определение функции, описывающей дифференциальное уравнение
def fun(x, y):
    return np.vstack((y[1], 6*x**2 + 1 + 4*y[1] - y[0]))

# Определение граничных условий
def bc(ya, yb):
    return np.array([ya[0], yb[0] + yb[1] + 3.5625])

# Задание начального приближения
x = np.linspace(0, 1, 10000)
y = np.zeros((2, x.size))

# Решение дифференциального уравнения с граничными условиями
sol = solve_bvp(fun, bc, x, y)

# Нарисовать график решения
plt.plot(sol.x, sol.y[0], label='u(x)')
plt.xlabel('x')
plt.ylabel('u(x)')
plt.legend()
plt.grid()
plt.show()
```
